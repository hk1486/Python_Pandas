```python
import pandas
from matplotlib import pyplot
```

```python
df = pandas.read_excel('bugs_2020년.xls',sheet_name=None)
df
```

```python
df_all = pandas.concat(df)
df_all
```

```python
song = df_all['title'] == 'All I Want for Christmas Is You'
df_song = df_all[song]
df_song
```

```python
date = []
for i in range(len(df_song.index)):
    date.append(df_song.index[i][0])
df_song.insert(2,'date',date)
```

```python
df_song.set_index('rank',inplace=True)
```

```python
df_song['date'] = pandas.to_datetime(df_song['date'])
```

```python
pyplot.figure(figsize = (15,7.5))
pyplot.plot(df_song['date'], df_song.index)
pyplot.ylim((100,0))
pyplot.legend(labels = ['2020'], loc = 'best')
```
