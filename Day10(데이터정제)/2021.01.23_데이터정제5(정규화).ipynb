```python
import pandas
from numpy import nan
```

```python
df = pandas.read_csv('auto-mpg.csv', header=None) 
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
              'model year', 'origin', 'name']

df.head(5)
```

```python
df['horsepower'].replace('?',nan,inplace=True)
df.dropna(subset=['horsepower'],axis=0,inplace=True)
df['horsepower'] = df['horsepower'].astype('float')
```

```python
df['model year'] = df['model year'] + 1900
```

## 변수들의 상대적인 크기 차이를 제거하기 위해 정규화

### 1. 각 값들을 최대값으로 나눠줌

```python
max = df['horsepower'].max()
min = df['horsepower'].min()

hp_rate = df['horsepower'] / max
df.insert(4,'hp_rate',hp_rate)
df
```

### 2. 값 - 최소값 / 최대값 - 최소값 으로 정규화

```python
min_x = df['horsepower'] - min
min_max = max - min
hp_rate2 = min_x / min_max
df.insert(5,'hp_rate2',hp_rate2)
```
