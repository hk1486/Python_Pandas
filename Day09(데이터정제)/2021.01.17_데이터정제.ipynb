```python
import seaborn
from numpy import nan
```

## 데이터를 불러와서 먼저 살펴보기

```python
df = seaborn.load_dataset('titanic')
df.info()
```

## nan값들을 찾아보기

```python
# 해당 열의 값들을 볼 수 있음
df['deck'].values
```

```python
# 해당 열의 값들의 수를 세어줌, dropna : nan의 수를 세어주는 함수
df['deck'].value_counts(dropna = False)
```

```python
# nan 데이터를 찾아줌 : isnull()
df.isnull().sum()
```

```python
df.head()
```

## 열의 nan값이 200개를 넘어가면 그 열은 Drop 해주기

```python
ndf = df.copy()
for column in df.columns:
    cls, cnt = column, df[column].isnull().sum()
    if cnt >= 200:
        ndf.drop(cls, axis = 1,inplace=True)
ndf
```

## nan값들 삭제하기

```python
df.dropna() # nan값이 하나라도 있으면 행 전체를 지워버림
```

```python
df.dropna(axis =1) # nan값이 하나라도 있는 컬럼들을 다 지워버림
```

```python
ndf = df.copy()
ndf.dropna(axis = 1, thresh = 500, inplace=True) # nan값의 수가 thresh 값을 넘어가 버리면 해당 열 혹은 행을 지워버림
ndf
```

## 특정열에서 nan값이 있는 행들을 지워버리기
- dropna(subset=['칼럼명'])
- how = 'any' : nan값이 하나라도 존재하면 지워줘
- how = 'all' : 모든 데이터가 nan값일 경우에만 지워줘 

```python
ndf.dropna(subset=['age','embarked','embark_town'],how='any')
```

## nan값을 평균값으로 치환하기(분석에 용이)

```python
mean_age = df['age'].mean() # 평균값 알아보기
df['age'] = df['age'].replace(nan, mean_age) # 평균으로 치환

df['age'] = list(map(int, df['age'])) # 나이 정수값으로 바꿔주기
df['age']
```

```python
df[['embarked','embark_town']].isnull().sum() # 널값 개수를 세본다 - 2개

df.dropna(subset=['embarked','embark_town'],axis = 0, how='any') # 2개 지워줌
```

## nan값을 가장 많이 나타나는 값으로 바꾸기

```python
most_freq = df['embark_town'].value_counts().index[0] # 어떤값이 가장 많이 나타나는지 알아보기
```

```python
df['embark_town'] = df['embark_town'].replace(nan,most_freq) # 그 값으로 치환
df['embark_town'].isnull().sum()
```
