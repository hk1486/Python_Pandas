```python
import pandas
from matplotlib import pyplot, font_manager, rc
from numpy import nan
import seaborn
```

```python
font_path = 'malgun.ttf'
font_name = font_manager.FontProperties(fname = font_path).get_name()
rc('font', family = font_name)
```

## 타이타닉 데이터 가져오기

```python
titanic = seaborn.load_dataset('titanic')
titanic.head()
```

```python
# 스타일 테마 설정
seaborn.set_style('whitegrid')
```

## 범주형 데이터의 산점도 그리기
- 나이와 등급에 따라

```python
figure = pyplot.figure(figsize=(15,5))
ax1 = figure.add_subplot(1,2,1)
ax2 = figure.add_subplot(1,2,2)

# 이산형 변수의 분포 - 데이터 분산 미고려(중복 표시)
seaborn.stripplot(x='class', y = 'age', data=titanic,
                 ax=ax1) # ax 객체 1번째 그래프

# 이산형 변수의 분포 - 데이터 분산 고려(중복 표시 x)
seaborn.swarmplot(x='class', y = 'age', data=titanic,
                 ax=ax2) # ax 객체 2번째 그래프

# 차트 제목 표시
ax1.set_title('Strip Plot')
ax2.set_title('Swarm Plot')

pyplot.show()
```

## 조인트 그래프
- 나이와 요금에 따라

```python
# 조인트 그래프 - 산점도(기본값)
j1 = seaborn.jointplot(x='fare', y='age', data=titanic)

# 조인트 그래프 - 회귀선
j2 = seaborn.jointplot(x='fare', y='age', data=titanic, kind = 'reg')

# 조인트 그래프 - 육각 그래프
j3 = seaborn.jointplot(x='fare', y='age', data=titanic, kind = 'hex')

# 조인트 그래프 - 커널 밀집 그래프
j4 = seaborn.jointplot(x='fare', y='age', data=titanic, kind = 'kde')

# 차트 제목 표시
j1.fig.suptitle('titanic fare - scatter', size= 15)
j2.fig.suptitle('titanic fare - reg', size= 15)
j3.fig.suptitle('titanic fare - hex', size= 15)
j4.fig.suptitle('titanic fare - kde', size= 15)

pyplot.show()
```
