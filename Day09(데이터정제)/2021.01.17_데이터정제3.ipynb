```python
import pandas
from numpy import nan
```

```python
df = pandas.read_csv('auto-mpg.csv', header=None) 
df.columns = ['mpg', 'cylinders', 'displacement', 'horsepower', 'weight', 'acceleration',
              'model year', 'origin', 'name']

df.head()
```

## 단위환산해서 새로운 컬럼 만들기

```python
# 갤런을 리터로 변환
# df['kpl'] = df['mpg'] * (1.60934/3.7854) 맨뒤에 컬럼이 생성됨

kpl = df['mpg'] * (1.60934/3.7854)
# df.insert(index_col, column_name, data(Series)) 내가 원하는 순서에 컬럼 생성하기

df.insert(1, 'kpl', kpl)
df
```

## 자료형 변환

```python
df['horsepower'].unique() # 열의 고유값 확인

df['horsepower'] = df['horsepower'].replace('?',nan) # '?' 값을 nan으로 바꿔주기

df['horsepower'].astype('float') # 데이터타입을 실수형으로 바꿔주기

df[['kpl','horsepower']].plot(kind='scatter', x = 'horsepower', y = 'kpl')
```
