```python
import pandas
from numpy import nan
```

```python
df = pandas.read_excel('naver_stock.xls')
df.head()
```

```python
df['new_date'] = pandas.to_datetime(df['date'])
df
```

## 연,월,일 정보만 따로 추출가능

```python
df['year'] = df['new_date'].dt.year
df['month'] =df['new_date'].dt.month
df['day'] =df['new_date'].dt.day
ndf = df.set_index('new_date')
ndf
# ndf = df.set_index(['year','month','day'])
# ndf.loc[(2021, 1)].loc[21:15]
```

## 원하는 날짜로 행 추출하기

```python
ndf.loc['2021-01-22']
ndf.loc['2021-01', ['start','max']]
```

## datetime 자료형으로 바꿔주기 위해 lambda를 사용해서 '-'로 바꿔준다

```python
df['date'] = df['date'].apply(lambda x: x.replace('.','-'))
df
```

## datetime 으로 바꿔주기

```python
new_date = pandas.to_datetime(df['date'])
df.insert(1, 'new_date', new_date)
df[['date','new_date']].info()
```

```python
df.set_index('new_date',inplace=True)
df
```

## 타임스탬프를 Period 데이터로 변환

```python
dates = ['2019-01-01','2020-03-01','2021-06-01']
to_dates = pandas.to_datetime(dates)
to_dates
```

## 주기를 'D' '일' 로 정해줌

```python
to_dates.to_period(freq='D')
```
