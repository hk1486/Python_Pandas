# Googlemaps 와 Folium

```python
import folium
import googlemaps
import pandas

# 구글 키값
my_key = 'input_key'
```

## 구글맵 활용하여 위치 데이터 불러오기

```python
maps = googlemaps.Client(key=my_key)
response = maps.geocode('송림빌딩')[0]

# 키값으로 위도, 경도 찾아서 변수에 저장
lat = response['geometry']['location']['lat']
lng = response['geometry']['location']['lng']
```

## 위도,경도를 찍어서 맵 불러오기

```python
seoul_map = folium.Map(location=[lat, lng], zoom_start=16)

seoul_map2 = folium.Map(location=[lat, lng], zoom_start=16,
              tiles = 'Stamen Terrain') # 지도 스타일 지정

seoul_map3 = folium.Map(location=[lat, lng], zoom_start=16,
              tiles = 'Stamen Toner') # 지도 스타일 지정

seoul_map2
# seoul_map.save('seoul_map.html')
```

## 대학교 위치 파일 불러와서 지도에 마킹하기

```python
df = pandas.read_excel('서울지역 대학교 위치.xls',header = 0)
df.columns = ['name','lat','lng']
df = df.set_index('name')

for name in df.index:
    lat = df.loc[name]['lat']
    lng = df.loc[name]['lng']
    folium.CircleMarker([lat,lng],
                        radius = 10, # 원의 반지름
                        color = 'brown', # 원의 둘레 색상
                        fill = True, # 원을 채우다
                        fill_color='coral', # 원을 채우는 색
                        fill_opacity=0.7, # 투명도
                        popup=name).add_to(seoul_map)
seoul_map
```

```python
# 지도 객체 생성
seoul_map = folium.Map(location = [37.55, 126.98], zoom_start=11.5)

# 지도에 마커 찍기
folium.Marker([lat,lng],popup=name).add_to(seoul_map)
```
