```python
import pandas as pd

dic_data = {'이름': ['서준','우현','인아'],
            '수학': [90,80,70],
            '영어': [98,89,95],
            '음악': [85,95,100],
            '체육': [100,90,90]}
df = pd.DataFrame(dic_data)
df
```

```python
df.set_index('이름',inplace=True)
print(df)
```

```python
df.loc['서준']['체육']=90
print(df)
```

```python
df.loc['서준',['음악','체육']] = 50
df.loc['서준',['음악','체육']] = 100, 50 # 파이썬에선 숫자를 콤마로 나열할시 튜플로 저장됨
df
```

```python
df2 = df.transpose() # 행과 열 바꿔줌
df2
```
